Ce choix de problématique a été fait car Toulouse est la ville où il y a le plus de données et le modèle prédictif en sera sûrement plus performant que si c'était une autre ville comme Paris où d'un quartier à l'autre, il pourrait y avoir des variations de prix énormes

# 3 Preprocessing

## 3.1 Preprocessing: Toulouse

On filtre notre dataframe sur Toulouse en faisant bien attention à prendre toutes les lignes d'une mutation ayant au moins une ligne dans la ville de Toulouse

```python
df1=df.loc[np.unique(df[df['nom_commune'] == 'Toulouse'].index.tolist())]
```

```python
#On vérifie les communes de notre nouveau dataframe df1
df1['nom_commune'].value_counts(dropna=False)
```

On voit bien qu'il y a des mutations qui concernent plusieurs biens (dont au moins un bien dans Toulouse et un dans une autre ville). Pour obtenir un modèle performant, nous allons exclure ces cas à la marge.

```python
indexaexclure=np.unique(df1[df1['nom_commune'] != 'Toulouse'].index.tolist())
```

```python
df2=df1.drop(indexaexclure,0)
```

```python
#On vérifie les communes de notre nouveau dataframe df2
df2['nom_commune'].value_counts(dropna=False)
```

## 3.2 Preprocessing: Appartements et maisons

On ne traitera que les mutations de biens qui comprennent au moins un appartement ou une maison

```python
df3=df2.loc[np.unique(df2[(df2['type_local'] == 'Appartement') | (df2['type_local'] == 'Maison')].index.tolist())]
```

## 3.3 Preprocessing : Gestion des valeurs manquantes

```python
df3.isna().sum()
#On doit décider des colonnes à drop
```

### 3.3.1 Traitement des valeurs manquantes (de manière grossière)

Nous allons supprimer les colonnes non exploitables (qui contiennent trop de valeurs vides) et les données de localisation ne seront pas traités ce qui veut dire qu'un bien en centre ville sera considéré de la même manière qu'un bien en périphérie

```python
df4= df3.drop(['numero_disposition',
               'adresse_numero',
               'adresse_suffixe',
               'adresse_nom_voie',
               'adresse_code_voie',
               'code_postal',
               'code_commune',
               'nom_commune',
               'code_departement',
               'ancien_code_commune',
               'ancien_nom_commune',
               'id_parcelle',
               'ancien_id_parcelle',
               'numero_volume',
               'lot1_numero',
               'lot1_surface_carrez',
               'lot2_numero',
               'lot2_surface_carrez',
               'lot3_numero',
               'lot3_surface_carrez',
               'lot4_numero',
               'lot4_surface_carrez',
               'lot5_numero',
               'lot5_surface_carrez',
               'nombre_lots',
               'longitude',
               'latitude'],axis= 1)
```

### 3.3.2 Traitement des valeurs manquantes (de manière détaillée)

```python
df4.isna().sum()
```

#### 3.3.2.1 Traitement des valeur_fonciere manquantes

Il est clair que nous ne pourrons pas travailler avec les lignes pour lesquelles la valeur foncière qui sera notre cible n'est pas présente

```python
#suppression des lignes pour lesquelles la valeur foncière est absente
df5=df4.drop(np.unique(df4[df4['valeur_fonciere'].isna()].index.tolist()),0)
```

#### 3.3.2.2 Traitement des type_local manquants

```python
df5[(df5['type_local'].isna())]
```

Les type_local vides semblent être des terrains, en effet la dernière colonne surface_terrain est remplie. Vérifions cette hypothèse

```python
df5[(df5['type_local'].isna()) & (df5['surface_terrain'].isna())]
```

On s'aperçoit que les lignes dont le type_local est vide et que la surface_terrain est vide sont toutes des mutations en VEFA. On décide donc de retirer les id_mutation sont dans ce cas là de notre dataframe. Peut-être même qu'il faudrait retirer tous les VEFA pour améliorer le modèle mais pour le moment on laisse les autres VEFA.

```python
indexaexclure=np.unique(df5[(df5['type_local'].isna()) & (df5['surface_terrain'].isna())].index.tolist())
df6=df5.drop(indexaexclure,0)
df6
```

```python
#On vérifie que les type_local vides sont bien des assimilables à des terrains et leur surface n'est pas vide
df6[df6['type_local'].isna()].groupby(['nature_culture'])['nature_culture'].count()
```

```python
#On va remplacer les type_local vide par "Terrain"
df6['type_local']=df6['type_local'].fillna('Terrain')
```

```python
df6.isna().sum()
```

### 3.3.3 Suppression des colonnes inutiles

Nous n'avons à présent plus besoin de certaines colonnes du dataframe

```python
df7= df6.drop(['code_type_local',
               'code_nature_culture',
               'nature_culture',
               'code_nature_culture_speciale',
               'nature_culture_speciale'],axis= 1)
```

### 3.3.4 Traitement des types_local : dépendances

```python
df7[(df7['type_local'] == 'Dépendance')].isna().sum()
```

Les dépendances n'ont normalement pas de surface terrain, comme on ne comprend pas pourquoi, on élimine du dataframe les cas où justement les dépendances ont une surface terrain

```python
indexaexclure=np.unique(df7[(df7['type_local'] == 'Dépendance') & (df7['surface_terrain'].notna())].index.tolist())
df8=df7.drop(indexaexclure,0)
df8
```

Les dépendances ont leur nombre_pieces_principales à 0, il ne devrait pas y avoir de valeurs vides

```python
condition = (df8['type_local'] == 'Dépendance')
df8.loc[condition,'nombre_pieces_principales'] = df8.loc[condition,'nombre_pieces_principales'].fillna(0)
```

### 3.3.5 Traitement des types_local : appartement

Les appartements n'ont normalement pas de surface terrain, comme on ne comprend pas pourquoi, on élimine du dataframe les cas où justement les appartements ont une surface terrain

```python
indexaexclure=np.unique(df8[(df8['type_local'] == 'Appartement') & (df8['surface_terrain'].notna())].index.tolist())
df9=df8.drop(indexaexclure,0)
df9
```

De plus les surfaces terrain des appartements devraient être à 0, on remplace donc les surface_terrain vides des appartements par 0

```python
condition = (df9['type_local'] == 'Appartement')
df9.loc[condition,'surface_terrain'] = df9.loc[condition,'surface_terrain'].fillna(0)
```

```python
df9.isna().sum()
```

### 3.3.6 Traitement des types_local : terrain

```python
df9[(df9['type_local'] == 'Terrain')].shape
```

```python
df9[(df9['type_local'] == 'Terrain')].isna().sum()
```

Les terrains n'ont pas de surface_reelle_bati ni de nombre_pieces_principales ce qui est logique, on va passer les vides à 0

```python
condition = (df9['type_local'] == 'Terrain')
df9.loc[condition,'surface_reelle_bati'] = df9.loc[condition,'surface_reelle_bati'].fillna(0)
df9.loc[condition,'nombre_pieces_principales'] = df9.loc[condition,'nombre_pieces_principales'].fillna(0)
```

```python
df9.isna().sum()
```

### 3.3.7 Traitement des types_local : appartement et maison sans surface_reelle_bati

Si on ne connait pas la surface_reelle_bati d'un appartement ou d'une maison, on ne peut clairement pas estimer le prix

```python
indexaexclure=np.unique(df9[((df9['type_local'] == 'Appartement') | (df9['type_local'] == 'Maison')) & (df9['surface_reelle_bati'].isna())].index.tolist())
df10=df9.drop(indexaexclure,0)
df10
```

### 3.3.8 Non traitement des types_local : Local industriel. commercial ou assimilé

```python
indexaexclure=np.unique(df10[(df10['type_local'] == 'Local industriel. commercial ou assimilé')].index.tolist())
df11=df10.drop(indexaexclure,0)
df11
```

### 3.3.9 Traitement des types_local : Dépendance - surface_reelle_bati - surface_terrain

Les dépendances doivent avoir leurs deux surfaces à 0

```python
condition = (df11['type_local'] == 'Dépendance')
df11.loc[condition,'surface_reelle_bati'] = df11.loc[condition,'surface_reelle_bati'].fillna(0)
df11.loc[condition,'surface_terrain'] = df11.loc[condition,'surface_terrain'].fillna(0)
```

```python
df11
```

### 3.3.10 Traitement des types_local : Maison - surface_terrain

Certaines maisons n'ont pas de terrain, on converti les valeurs vide des terrains des maisons en 0

```python
condition = (df11['type_local'] == 'Maison')
df11.loc[condition,'surface_terrain'] = df11.loc[condition,'surface_terrain'].fillna(0)
```

```python
df11.isna().sum()
```

### 3.3.11 Traitement des nature_mutation : Autre que Vente et VEFA

```python
df11['nature_mutation'].value_counts()
```

Les Adjudication, Echange et Vente terrain à bâtir peuvent être très décorrélées du marché immobilier, nous allons les enlever de notre dataframe

```python
indexaexclure=np.unique(df11[(df11['nature_mutation'] == 'Adjudication') | (df11['nature_mutation'] == 'Echange') | (df11['nature_mutation']== 'Vente terrain à bâtir')].index.tolist())
df12=df11.drop(indexaexclure,0)
df12
```

# 5 Finalisation du dataframe pour la Modèlisation

## 5.1 Regroupement des id_mutation

Nous allons faire en sorte que chaque ligne contienne une transaction c'est-à-dire qu'on veut que l'id_mutation soit unique

```python
df13=pd.get_dummies(df12, columns=['nature_mutation', 'type_local'], prefix=['nature_mutation', 'type_local'])
df13
```

```python
functions_to_apply = {
    'valeur_fonciere':[np.mean],
    'surface_reelle_bati' : ['sum'],
    'nombre_pieces_principales' : ['sum'],
    'surface_terrain' : ['sum'],
    'nature_mutation_Vente' : ['sum'],
    "nature_mutation_Vente en l'état futur d'achèvement" : ['sum'],
    'type_local_Appartement' : ['sum'],
    'type_local_Dépendance' : ['sum'],
    'type_local_Maison' : ['sum'],
    'type_local_Terrain' : ['sum']
    }
df14=df13.groupby(['id_mutation']).agg(functions_to_apply)
df14.head()
```

## 5.2 Gestion des ventes de plusieurs appartements et maison dans une même transaction

Les mutations qui concernent plus d'une (maison + appartement) ne seront pas prises en compte car les immeubles de rapport et des ventes en lot sont quelque peu différents d'une vente unique (un rabais est souvent appliqué).

```python
df14['AppartEtMaison']=df14['type_local_Appartement','sum'] + df14['type_local_Maison','sum'] 
```

```python
df15=df14[df14['AppartEtMaison']==1]
df15=df15.drop(['AppartEtMaison'], axis = 1)
df15.columns = ['valeur_fonciere',
                'surface_reelle_bati',
                'nombre_pieces_principales',
                'surface_terrain',
                'nature_mutation_Vente',
                "nature_mutation_Vente en l'état futur d'achèvement",
                'type_local_Appartement',
                'type_local_Dépendance',
                'type_local_Maison',
                'type_local_Terrain']
df15
```

## 7.1 Modélisation: régression linéaire simple

La valeur foncière étant le plus corrélée à la surface bâtie (après le nombre_pieces_principales mais on ne va pas utiliser ça), on va essayer de faire un premier modèle de régression linéaire simple et voir le résultat

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data = df19.drop('valeur_fonciere', axis=1)
target = df19['valeur_fonciere']
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=101)
df_comparaison_modeles = pd.DataFrame(columns=['nom_modele','score_train','score_test','rmse_train','rmse_test','mae_test','commentaires'])
pd.set_option('display.max_colwidth', None)
```

## 7.1.1 Sans normalisation les données

```python
#Est-ce qu'il faut normaliser? à essayer
#scaler = preprocessing.StandardScaler()
#X_train[X_train.columns] = pd.DataFrame(scaler.fit_transform(X_train), index=X_train.index)
#X_test[X_test.columns] = pd.DataFrame(scaler.transform(X_test), index=X_test.index)
```

```python
modeleLR = LinearRegression()
modeleLR.fit(X_train[['surface_reelle_bati']], y_train)
modele=modeleLR
y_pred_train = modele.predict(X_train[['surface_reelle_bati']])
y_pred_test = modele.predict(X_test[['surface_reelle_bati']])
```

```python
score_train=modele.score(X_train[['surface_reelle_bati']], y_train)
score_test=modele.score(X_test[['surface_reelle_bati']], y_test)
rmse_train=np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_test=np.sqrt(mean_squared_error(y_test, y_pred_test))
mae_test=mean_absolute_error(y_test, y_pred_test)
print('score_train :',score_train)
print('score_test :', score_test)
print('rmse_train :', rmse_train)
print('rmse_test : ', rmse_test)
print('mae_test : ', mae_test)

```

On retiendra de ce projet la difficulté du pre-processing.  
En effet on est parti d'un jeu de données de 17,2 millions de mutations.  
On a filtré sur les mutations comprennant au moins un appartement ou une maison (+les éventuels dépendances: jardins, cave...) sur Toulouse.
On a supprimé les colonnes du dataset non pertinentes au projet.  
Un gros travail a été effectué pour gérer les valeurs manquantes et les valeurs aberrantes.  
Seuls les mutations de type "vente" et "VEFA" ont été gardés, les autres types de mutation seraient probablement trop décorrélées du marché.  
Les mutations avec un même ID ont été regroupées, c'est à dire qu'on a fait en sorte qu'une ligne du nouveau dataframe représente une transaction (qui comprend donc un ou plusieurs biens mutés).  
Les transactions qui comprennent plus d'un appartement/maison ont été retirés du jeu de données car les ventes en lot ne sont pas représentatif du marché immobilier.  
Les valeurs extrêmes ont été retirées (valeur foncière inférieurs à 40k€ et supérieure à 1 million par exemple).
